```python
df.isna().sum()
```

```python
# Removing columns I won't be needing.
df.drop(columns = ['Rating', 'Headquarters', 'Founded', 'Sector', 'Revenue', 'Competitors', 'Industry'], inplace = True)
```

```python
df
```

```python
# chaecking for duplicates
df.duplicated().sum()
```

```python
# checking for columns that contain invalid value

columns_with_invalid = df[(df == '-1').any(axis=1)]

columns_with_invalid
```

```python
# Dropping rows that contains invalid

mask = df['Size'] == '-1'
df = df.drop(df[mask].index)
```

```python
# Checking again for columns that contains invalid

columns_with_invalid = df[(df == '-1').any(axis=1)]

columns_with_invalid
```
